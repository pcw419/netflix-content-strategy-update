### Initial Data Inspection Summary

- Both movie and TV datasets contain approximately 16,000 rows.
- Movies include financial variables such as budget and revenue, while TV shows do not.
- Several columns (e.g., director, cast, budget) contain a large number of missing values.
- The `duration` column has different meanings for movies (minutes) and TV shows (seasons).

```python
# Handle missing values in movies dataset
# Fill missing categorical values with 'Unknown'

movies_clean = movies_df.copy()

movies_clean['director'] = movies_clean['director'].fillna('Unknown')
movies_clean['cast'] = movies_clean['cast'].fillna('Unknown')
movies_clean['country'] = movies_clean['country'].fillna('Unknown')
movies_clean['genres'] = movies_clean['genres'].fillna('Unknown')
movies_clean['description'] = movies_clean['description'].fillna('')

# Drop duration column (not consistent across datasets)
movies_clean = movies_clean.drop(columns=['duration']) 
```

```python
# Handle missing values in TV shows dataset
# Fill missing categorical values with 'Unknown'

tv_clean = tv_df.copy()

tv_clean['director'] = tv_clean['director'].fillna('Unknown')
tv_clean['cast'] = tv_clean['cast'].fillna('Unknown')
tv_clean['country'] = tv_clean['country'].fillna('Unknown')
tv_clean['genres'] = tv_clean['genres'].fillna('Unknown')
tv_clean['description'] = tv_clean['description'].fillna('')
```

```python
# DataFrame.copy() creates an independent copy of the original dataset to safely perform data cleaning without mutating the raw data.
```

```python
# Convert date_added to datetime format
movies_clean['date_added'] = pd.to_datetime(movies_clean['date_added'])
tv_clean['date_added'] = pd.to_datetime(tv_clean['date_added'])
```
